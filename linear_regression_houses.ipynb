# Example of Linear Regression Prediction with house prices

```python
# Import pandas library and the data

import pandas as pd

data = pd.read_csv("housing_data.csv")

data.info
```

```python
# Check the data

data.head()
```

```python
# Check the column with problems

data["ocean_proximity"].value_counts()
```

```python
# Check if there are data in all the rows

data.info()
```

```python
# Check more the data

data.describe()
```

```python
# Plot the data

data.hist(figsize=(15,8), bins=30, edgecolor="black")
```

```python
# Plot the location of the houses with the price

import seaborn as sb

sb.scatterplot(x="latitude", y="longitude", data=data,
               hue="median_house_value", palette="coolwarm")
```

```python
# Clean empty rows

data_na = data.dropna()

data_na.info()
```

```python
# Clean non numeric colum with dummies

dummies = pd.get_dummies(data_na["ocean_proximity"], dtype=int)

data_na = data_na.join(dummies)

data_na = data_na.drop(["ocean_proximity"], axis=1)

data_na
```

```python
# Check correlation in a graph (1 -> Total correlation, -1 -> Total uncorrelation)

sb.set_theme(rc={'figure.figsize':(14,8)})
sb.heatmap(data_na.corr(), annot=True, cmap="YlGnBu")
```

```python
# Check correlation of the price in a list

data_na.corr()["median_house_value"].sort_values(ascending=False)
```

```python
# The median income is the data with more correlation to the house price

sb.scatterplot(x=data_na["median_house_value"], y=data_na["median_income"])
```

```python
# To prepare the data to train the model: Characteristics and label

X = data_na.drop(["median_house_value"], axis=1)
y = data_na["median_house_value"]
```

```python
# Set of training and set to test
from sklearn.model_selection import train_test_split

X_training, X_test, y_training, y_test = train_test_split(X, y, test_size=.2)
```

```python
# Train the model

from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(X_training, y_training)
```

```python
# Test vs the test set

predictions = model.predict(X_test)

checker = {"predictions": predictions, "test value": y_test}
pd.DataFrame(checker)
```

```python
# Check score

print(model.score(X_training, y_training))
print(model.score(X_test, y_test))
```

```python
# Check mean squared error

from sklearn.metrics import mean_squared_error
import numpy as np

mse = mean_squared_error(predictions, y_test)

np.sqrt(mse)
```
