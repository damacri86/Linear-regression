# Example of Linear Regression Prediction with house prices

```python
# Import pandas library and the data

import pandas as pd

data = pd.read_csv("housing_data.csv")

data.info
```

```python
# Check the data

data.head()
```

```python
# Check the column with problems

data["ocean_proximity"].value_counts()
```

```python
# Check if there are data in all the rows

data.info()
```

```python
# Check more the data

data.describe()
```

```python
# Plot the data

data.hist(figsize=(15,8), bins=30, edgecolor="black")
```

```python
# Plot the location of the houses with the price

import seaborn as sb

sb.scatterplot(x="latitude", y="longitude", data=data,
               hue="median_house_value", palette="coolwarm")
```

```python
# Clean empty rows

data_na = data.dropna()

data_na.info()
```

```python
# Clean non numeric colum with dummies

dummies = pd.get_dummies(data_na["ocean_proximity"], dtype=int)

data_na = data_na.join(dummies)

data_na = data_na.drop(["ocean_proximity"], axis=1)

data_na
```

```python
# Check correlation in a graph (1 -> Total correlation, -1 -> Total uncorrelation)

sb.set_theme(rc={'figure.figsize':(14,8)})
sb.heatmap(data_na.corr(), annot=True, cmap="YlGnBu")
```

```python
# Check correlation of the price in a list

data_na.corr()["median_house_value"].sort_values(ascending=False)
```
