# Example of Linear Regression Prediction for weight and height

```python
# Import pandas library

import pandas as pd

data = pd.read_csv("height_weight_data.csv")

data.info
```

```python
# Check the data

data.head()
```

```python
# Represent the data

import seaborn as sb

sb.scatterplot(x="height", y="weight", data=data)
```

```python
# Reshape the data

X = data["height"]
y = data["weight"]

X_reshaped = X.values.reshape(-1,1)
y_reshaped = y.values.reshape(-1,1)
```

```python
# Train the model

from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(X_reshaped, y_reshaped)
```

```python
# Test the model

model.predict([[160]])
```

```python
# Check the score

model.score(X_reshaped, y_reshaped)
```
